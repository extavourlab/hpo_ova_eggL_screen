# Building the meta module network

```python
# Create the network composed of all 4 modules aggregated
subG = nx.compose(ova_module_G, fec_module_G)
subG = nx.compose(subG, xRNAi_module_G)
subG = nx.compose(subG, core_module_G)
```

```python
# Extract the locations in modules for all genes in the MetaNetwork
for node in subG.nodes():
    # Define the location in modules
    # Ova / HipoEggL / Eggl 
    key = ['0','0','0']
    if node in ova_module:
        key[0] = '1'
    if node in fec_module:
        key[1] = '1'
    if node in xRNAi_module:
        key[2] = '1'
    if node in core_module:
        key[0] = '1'
        key[1] = '1'
        key[2] = '1'
    subG.nodes[node]['overlap'] = ''.join(key)
```

```python
# Extract the Z_scores for all the nodes
scores_ova = {}
scores_fec = {}
scores_xRNAi = {}
scores = {}
for node in subG.nodes():
    Z = []
    if node in mean_xRNAi_gene['FbID'].unique():
        z = mean_xRNAi_gene[mean_xRNAi_gene['FbID'] == node]['Z'].values[0]
        Z.append(z)
        scores_xRNAi[node] = z
    if node in mean_fec_gene['FbID'].unique():
        z = mean_fec_gene[mean_fec_gene['FbID'] == node]['Z'].values[0]
        Z.append(z)
        scores_fec[node] = z
    if node in mean_ova_gene['FbID'].unique():
        z = mean_ova_gene[mean_ova_gene['FbID'] == node]['Z'].values[0]
        Z.append(z)
        scores_ova[node] = z
    if len(Z) == 3:
        scores[node] = np.mean(Z)
```

```python
# Decorate the network by adding all the calculated fileds above
for node in subG.nodes():
    if node in scores:
        Z = scores[node]
        if node in scores_ova:
            Z_ova = scores_ova[node]
        else:
            Z_ova = 0.0
        if node in scores_fec:
            Z_fec = scores_fec[node]
        else:
            Z_fec = 0.0
        if node in scores_xRNAi:
            Z_xRNAi = scores_xRNAi[node]
        else:
            Z_xRNAi = 0.0
        subG.nodes[node]['Z'] = abs(float(Z))
        subG.nodes[node]['Z_ova'] = abs(float(Z_ova))
        subG.nodes[node]['Z_fec'] = abs(float(Z_fec))
        subG.nodes[node]['Z_xRNAi'] = abs(float(Z_xRNAi))
    else:
        subG.nodes[node]['Z'] = 0.0
        if node in scores_ova:
            subG.nodes[node]['Z_ova'] = abs(float(scores_ova[node]))
        else:
            subG.nodes[node]['Z_ova'] = 0.0
        
        subG.nodes[node]['Z_fec'] = 0.0
        subG.nodes[node]['Z_xRNAi'] = 0.0
    sigs = signaling[signaling['FbID'] == node]['Sig'].values
    for sig in signaling['Sig'].unique():
        if sig in sigs:
            subG.nodes[node][sig] = 1
        else:
            subG.nodes[node][sig] = 0
```

```python
# Adding the labels on nodes if they are receptor or ligand

for node in subG.nodes():
    if node in signaling_meta['FbID'].values:
        types = signaling_meta[signaling_meta['FbID']== node]['KeyMember'].values[0]
        if type(types) == str:
            if types == "Ligand":
                subG.nodes[node]["SigType"] = "Ligand"
            else:
                subG.nodes[node]["SigType"] = "Receptor"
        else:
            subG.nodes[node]["SigType"] = "Else"            
    else:
        subG.nodes[node]["SigType"] = "Else"
```

# Analysing Edge Density between different gene sets

```python
modules = ["001","100","010","111","011","101","110"]
modules_leg = ["EggL+","Ova+","HpoEggL+","Core","EggL+/HpoEggL+","Ova+/EggL+","Ova+/HpoEggL+"]

res = np.zeros((len(modules),len(modules)))
c_i = 0
for i in range(len(modules)):
    module_i = modules[i]
    nodes_i = [i for i in subG.nodes if subG.nodes[i]['overlap'] == module_i]
    connector_i = [i for i in subG.nodes if subG.nodes[i]['overlap'] == module_i and subG.nodes[i]['Seed'] == 0]

    for j in range(len(modules)):
        module_j = modules[j]
#         if module_i == module_j:
#             res[i][j] = np.nan
#         else:
        nodes_j = [i for i in subG.nodes if subG.nodes[i]['overlap'] == module_j]
        edges = 0
        for node_i in nodes_i:
            for node_j in nodes_j: 
                if subG.has_edge(node_i, node_j):
#                     print(node_i, node_j)
                    edges += 1
        if module_i == module_j:
            n = len(nodes_i)
            res[i][j] = edges/( (n**2 - n) / 2 )
        else:
            max_edge = len(nodes_i) * len(nodes_j)
            res[i][j] = edges/max_edge
    
df = pd.DataFrame(res, columns=modules_leg, index=modules_leg)
```

# Analyzing the connectivity between groups 
Here we will try to analyze if the different subgroups are more connected than we would expect at random. 

To do this, we can take the same subnetwork, and measure the edge density between groups like before. But each time, we randomly assign the genes in a different group. 
As the reviewer pointed out, we should try to run this only on the seed genes, avoiding the Connector genes due to potentially inflated connectivity. 

I would prefer to run both for thoroughness especially because it does not cost more to do so.

```python
import seaborn as sns
from tqdm import notebook as tqdm
```

```python
# Ok first let's make a random assignement in the network.

def make_randomly_assigned_network(network):
    # We need to match the node id with their groups in two arrays
    # So let's grab the nodes 
    nodes = list(network.nodes())
    # And then we take the grouping of this node
    groups = [network.nodes[node]['overlap'] for node in nodes]
    # now we are going to shuffle the groups, and reassign to a copy of the network those groups.
    np.random.shuffle(groups)
    # We make an offline copy of the starting network
    network = network.copy()
    # And then reassign the shullfed groups to the nodes so we conserve the number of nodes per group
    for node, group in zip(nodes, groups):
        network.nodes[node]['overlap'] = group
    return network
    
def filter_connectors(network):
    # So let's grab the nodes 
    nodes = list(network.nodes())
    # And filter for the seeds if we only want the seeds
    nodes = [node for node in nodes if 'Seed' in network.nodes[node]['Seed']]
    return network.subgraph(nodes)

def compute_edge_density(network):
    modules = ["001","100","010","111","011","101","110"]
    modules_leg = ["EggL+","Ova+","HpoEggL+","Core","EggL+/HpoEggL+","Ova+/EggL+","Ova+/HpoEggL+"]

    res = []
    for i in range(len(modules)):
        nodes_i = [node for node in network.nodes() if network.nodes[node]['overlap'] == modules[i]]
        for j in range(i, len(modules)):
            nodes_j = [node for node in network.nodes() if network.nodes[node]['overlap'] == modules[j]]
            tmp = [modules[i] + " - " + modules[j], modules_leg[i] + " - " + modules_leg[j]]
            edges = 0
            for node_i in nodes_i:
                for node_j in nodes_j: 
                    if network.has_edge(node_i, node_j):
                        edges += 1
            if modules[i] == modules[j]:
                n = len(nodes_i)
                tmp.append(edges/( (n**2 - n) / 2 ))
            else:
                max_edge = len(nodes_i) * len(nodes_j)
                tmp.append(edges/max_edge)
            res.append(tmp)

    df = pd.DataFrame(res, columns=["module_i-j", "module_leg_i-j", "Density"])
    return df

def compute_z_scores(random, meta):
    modules = meta['module_i-j'].unique()
    modules_index = ["001","100","010","111","011","101","110"]
    modules_leg = ["EggL+","Ova+","HpoEggL+","Core","EggL+/HpoEggL+","Ova+/EggL+","Ova+/HpoEggL+"]

    res = np.zeros((len(modules_index), len(modules_index)))
    for module in modules:
        module_i, module_j = module.split(' - ')
        i, j = modules_index.index(module_i), modules_index.index(module_j) 
        mu = random[random['module_i-j'] == module]['Density'].mean()
        std = random[random['module_i-j'] == module]['Density'].std()
        x = meta[meta['module_i-j'] == module]['Density'].mean()
        res[j][i] = (x-mu)/std
    return pd.DataFrame(res, columns=modules_leg, index=modules_leg)
```
